## HTTP
    #### Hyper Text Transfer Protocol
* Request, Response Cycle
    * Request: method, identifier, [data]
    * Response: status code, [data]
    * Both may have arbitrary key-values as well
    * Nomenclature : 
        * Response Head refers to status code and key-values
        * Response Body refers to the (optional!) data

## HTTP Request

#### GET /anything HTTP/1.1
#### Accept: \*/\*
#### Accept-Encoding: gzip, deflate
#### Connection: keep-alive
#### Host: httpbin.org
#### User-Agent: HTTPie/0.9.8


### Another Http Request
#### POST /anything HTTP/1.1
#### Accept: application/json, \*/\*
#### Content-Type: application/json
#### {
####        "hello" : "there" 
####      }

## HTTP Response

#### HTTP/1.1 200 OK
#### Access-Control-Allow-Credentials: true
#### ACCess-Control-Allow-Origin: *
#### Content-Length: 344
#### Content-Type: application/json

#### { ... }

## HTTP Request Methods

* Idempotent (safe):
    * GET (Head + Body)
    * HEAD (Head only)
    * OPTIONS (Allowed Methods)
* Data-Changing : 
    * POST (New Data)
    * PUT (Update Data)
    * PATCH (Partial Update)
    * DELETE

### HTTP Response Codes
<ul>
    <li> 100s Informational response(4) </li>
    <li> 200s Success (10) </li>
    <li> 300s Redirection (9) </li>
