#### SQL
     MySQL and Sqlite3
     
     Database --> Tables --> Data
     
     4 Types of Languages:
     
         1. DDL
             Data Definition Language 
             Overall definition of your table is Schema
                 1. create
                     to create table/db or to create schema
                 2. drop
                     drop whole table/db with whole data
                 3. alter
                     table modify (add col, drop column, modify column)
                 4. truncate
                     drop whole data but not schema
                 5. rename
                 6. comment
         2. DML
             Data Manipulation Language
             1. insert
                 It is used to insert data into table
             2. update
                 It will modify the data 
             3. delete
                 Particular rows delete on the basis of condition
             4. select
                 Select the data 
         3. DCL
             Data Control Language
             1. Grant
                 Give permission
             2. Revoke
                 Take back permission
         4. TCL
             Transaction Control Language
             1. Commit
                 Permanently save
             2. Start
             3. Rollback
                 Getting back
             4. Savepoint

#### Keys

    1. Primary key
        A column which will be uniquely identify our rows
        There can be only 1 pk in your table
        It will not consists of duplicate values
        It cannot be null
    2. Unique Key
        A column will not consists of duplicate values 
        It can be null
    3. Foreign Key
    4. Candidate Key
        The keys which are eligible to be your primary keys
    5. Super Key
        When all columns combine to make a primary key
    6. Alternate Key
        The keys which cannot be your primary key
        eg --> Name, Stream, Course, Age 
    7. Composite Key
         When we combinely use 2 or more columns as primary key
