##### Top 10 skills and perks

```python
df['Skills and Perks']
```

```python
d={}
for i in df['Skills and Perks'].values:
    for j in eval(i):
        j = j.strip().lower()
        if j in d:
            d[j]+=1
        else:
            d[j]=1
```

```python
top_10 = dict(sorted(d.items(),key=lambda x: x[1], reverse=True)[:10])
```

```python
top_10.values()
```

```python
fig = plt.figure(facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")
sns.barplot(x=list(top_10.keys()), y=list(top_10.values()))
plt.xticks(rotation=90)
plt.tick_params(color="white", labelcolor="white")
plt.title("TOP SKILLS AND PERKS", fontsize=15, color="red")
plt.savefig("sp.png", bbox_inches="tight")
plt.show()
```
