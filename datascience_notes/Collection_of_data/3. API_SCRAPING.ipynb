```python
def scrap_movies():
    url = "https://www.timeout.com/london/film/the-100-best-bollywood-movies-the-list"
    resp = requests.get(url)
    if resp.status_code == 200:
        soup = BeautifulSoup(resp.text, 'html') # converting into html format

        # storing the movies in movies list to access it later
        movies = []
        for movie in soup.select("._h3_cuogz_1"):
            movies.append(movie.text.replace(".", " (").split(" (")[1][1:])
        return movies
    
    
def get_movie_detail(title):
    key = "e22bdd41"
    url = f"http://www.omdbapi.com/?apikey={key}&t={title}"
    resp = requests.get(url)
    if resp.status_code == 200:
        data = resp.json()
        key = ['Year', 'Rated', 'Released', 'Runtime', 'Genre', 'Director', 'imdbRating', 'Actors',
              'Plot', 'Country']
        values = []
        for k in key:
            values.append(data.get(k))
        return values 
    return "Invalid Url"
    
def main():
    key = ['Movie', 'Year', 'Rated', 'Released', 'Runtime', 'Genre', 'Director', 'imdbRating', 'Actors',
            'Plot', 'Country']
    data = {i:[] for i in key}
    movies = scrap_movies()
    for m in movies:
        val = get_movie_detail(m)
        val.insert(0, m)
        for k, v in zip(key, val):
            data[k].append(v)
    return data 

data = main()
```
