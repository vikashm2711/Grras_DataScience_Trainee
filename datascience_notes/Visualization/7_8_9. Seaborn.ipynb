```python
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'])

df.groupby(df['Dt_Customer'].datetime.month)['Income'].nlargest(10)['NumWebVisitsMonth'].value_counts()
```

```python
df.groupby("Income").max()['NumWebVisitsMonth'].sort_index(ascending=False).head(10)
```

```python
i = df.groupby("Income").max()['NumWebVisitsMonth'].sort_index().head(10).index
v =  df.groupby("Income").max()['NumWebVisitsMonth'].sort_index().head(10).values
```

```python
sns.lineplot(i, v, color='red', marker='*', ms=10, ls='--')
plt.grid()
plt.xticks(i)
plt.title("MINIMUM INCOME PEOPLE VISITES PER MONTH ON WEB", color='green', fontsize=20)
plt.show()
```

```python
i = df.groupby("Income").max()['NumWebVisitsMonth'].sort_index(ascending=False).head(10).index
v = df.groupby("Income").max()['NumWebVisitsMonth'].sort_index(ascending=False).head(10).values
sns.lineplot(i, v, color='red', marker='*', ms=10, ls='--')
plt.grid()
plt.xticks(rotation=90)
plt.title("MAXIMUM INCOME PEOPLE VISITES PER MONTH ON WEB", color='green', fontsize=20)
plt.show()
```

#### How many People above or equal to 60 years who are single and spending amount on wines greater than 200
#### How many People above or equal to 60 years who are married and spending amount on wines greater than 200

```python
df['Age'] = 2021 - df['Year_Birth']
```

```python
var1 = df[(df['Age'] >= 60) & (df['Marital_Status'] == 'Single') & (df['MntWines']>200)].__len__()
```

```python
var2 = df[(df['Age'] >= 60) & (df['Marital_Status'] == 'Married') & (df['MntWines']>200)].__len__()
```

```python
index = ['Single', 'Married']
values = [var1, var2]
sns.barplot(index, values)
plt.xlabel("MARITAL STATUS")
plt.ylabel("AMOUNT ON WINES GREATER THAN 200")
plt.show()
```
