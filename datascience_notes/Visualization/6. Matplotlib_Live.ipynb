```python
fig, ax = plt.subplots()
x = np.random.randint(1, 10, 10)
ax.plot(x, 'ro--')
def animatefunc(default):
    global x
    x = np.delete(x, 0)
    x = np.append(x, np.random.randint(1, 20))
    ax.plot(x, 'ro--')
    
f = FuncAnimation(fig, animatefunc)
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.random.randint(1, 10, 10)
ax.plot(x, 'ro--')
def animatefunc(default):
    global x
    ax.clear()
    x = np.delete(x, 0)
    x = np.append(x, np.random.randint(1, 20))
    ax.plot(x, 'ro--')
    
f = FuncAnimation(fig, animatefunc)
plt.show()
```

```python
cpu = []
from time import sleep
for i in range(10):
    sleep(1)
    cpu.append(psutil.cpu_percent())
```

```python
fig, ax = plt.subplots()
per = cpu.copy()
ax.plot(per, 'ro--')
def animatefunc(default):
    global per
    ax.clear()
    #sleep(1)
    per = np.delete(per, 0)
    per = np.append(per, psutil.cpu_percent())
    ax.plot(per, 'ro--')
    
f = FuncAnimation(fig, animatefunc)
plt.show()
```
