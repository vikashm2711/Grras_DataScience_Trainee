```python
fig = plt.figure(facecolor="grey")
ax1 = fig.add_axes([0, 0, 0.5, 0.5])
ax2 = fig.add_axes([0.6, 0, 0.5, 0.5])
ax3 = fig.add_axes([0, 0.6, 0.5, 0.5])
ax4 = fig.add_axes([0.6, 0.6, 0.5, 0.5])
i1 = data['Education'].value_counts().index 
v1 = data['Education'].value_counts().values
i2 = data['Year_Birth'].value_counts()[:10].index
v2 = data['Year_Birth'].value_counts()[:10].values
ax1.bar(i1, v1, color=[color_generate() for i in range(len(i1))])
ax2.plot(data['Income'], 'g*')
ax3.hist(data['Income'], color='grey', ec='k')
ax4.bar(i2, v2, color = [color_generate() for i in range(len(i2))])
for ax in [ax1, ax2, ax3, ax4]:
    ax.set_facecolor("grey")
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
fig = plt.figure(facecolor="black")
ax1 = fig.add_axes([0, 0, 0.5, 0.5])
ax2 = fig.add_axes([0.6, 0, 0.5, 0.5])
ax3 = fig.add_axes([0, 0.7, 0.5, 0.5])
ax4 = fig.add_axes([0.6, 0.7, 0.5, 0.5])
i1 = data['Education'].value_counts().index 
v1 = data['Education'].value_counts().values
i2 = data['Year_Birth'].value_counts()[:10].index
v2 = data['Year_Birth'].value_counts()[:10].values
ax1.bar(i1, v1, color=[color_generate() for i in range(len(i1))])
ax2.plot(data['Income'], 'g*')
ax3.hist(data['Income'], color='grey', ec='k')
ax4.bar(i2, v2, color = [color_generate() for i in range(len(i2))])
ax1.set_xticklabels(i1, rotation=45)
ax4.set_xticklabels(i2, rotation=40)
ax4.set_xticks(i2)
titles = ['Education Plot', 'Income Plot', 'Income Distribution', 'Top 10 Birth Year Customers']
axes = [ax1, ax2, ax3, ax4]
for i in range(len(axes)):
    axes[i].set_facecolor("black")
    axes[i].tick_params(labelcolor='white', labelsize=10)
    axes[i].set_title(titles[i], color="white", fontsize=17)
```
