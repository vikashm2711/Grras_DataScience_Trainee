```python
plt.plot(x, x**2, 'ko')
plt.xlabel("X", fontsize=15, color='#abc123')
plt.ylabel("$X^2$", fontsize=15, color="#abc123", fontstyle="italic")
```

```python
label = {'fontsize': 15, 'color': '#abc123', 'fontfamily': 'monospace'}
title = {'fontsize': 20, 'color': 'IndianRed', 'fontfamily': 'sans-serif'}
```

```python
plt.plot(x, x**2, 'ko')
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
plt.title("X Vs $X^2$", fontdict=title)
```

```python
plt.plot(x, x**2, 'ko')
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
plt.title("X Vs $X^2$", fontdict=title)
plt.xticks(range(1, 106, 5), color="#417441", fontsize=12)
plt.yticks(color="#417441", fontsize=12)
plt.grid()
plt.show()
```

```python
plt.plot(x, x**2, 'ko')
plt.xlabel("X", fontdict=label)
plt.ylabel("$X^2$", fontdict=label)
plt.title("X Vs $X^2$", fontdict=title)
plt.xticks(range(1, 106, 5))
plt.tick_params(size=12, color="blue", labelcolor="#417441", labelsize=12, width=10)
plt.show()
```

```python
x = np.arange(1, 50, 2)
```

```python
plt.plot(x, x, 'g^', label="X")
plt.plot(x, x**2, 'k*', label="$X^2$")
plt.xlabel("X", fontdict=label)
plt.ylabel("X/$X^2$", fontdict=label)
plt.title("X Vs X & $X^2$", fontdict=title)
plt.tick_params(size=12, color="blue", labelcolor="#417441", labelsize=12, width=10)
plt.legend()
plt.show()
```
