```python
def download(url, filename):
    try:
        resp = requests.get(url)
        if resp.status_code == 200:
            with open(filename, 'wb') as fp:
                fp.write(resp.content)
        else:
            return "Invalid URL"
    except Exception as error:
        return error
```
