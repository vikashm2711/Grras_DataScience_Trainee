```python
values = {}
col = data.split("\n")[0].split(",")
for i in data.split("\n")[1:-1]:
    #print(i.split(","))
    for k, v in zip(col, i.split(",")):    
        if k in values:
            values[k].append(v)  # appending in existing key
        else:
            values[k] = [v]  # creating new key
```
