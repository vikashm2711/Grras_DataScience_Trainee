```python
df1 = pd.DataFrame({
    "Id": np.arange(1, 11),
    "Name": np.array([chr(i) for i in range(65, 75)]),
    "Maths": np.random.randint(40, 100, 10)
})
```

```python
df2 = pd.DataFrame({
    "Id": np.arange(1, 11),
    "Science": np.random.randint(40, 100, 10),
    "English": np.random.randint(40, 100, 10)
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)  # inner join 
```

```python
df2.columns = ['Sid', 'Science', 'English']
```

```python
df2
```

```python
df1
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, left_on="Id", right_on="Sid")  # left --> df1, right --> df2
```

```python
df2.Sid = [1, 2, 3, 4, 5, 11, 12, 13, 14, 15]
```

```python
df2
```

```python
pd.merge(df1, df2, left_on="Id", right_on="Sid") 
```

```python
df2['Sid'] 
```

```python
pd.merge(df1, df2, how="outer", left_on="Id", right_on="Sid")  # outer join 
```

```python
pd.merge(df1, df2, how="left", left_on="Id", right_on="Sid")  # left join 
```

```python
pd.merge(df1, df2, how="right", left_on="Id", right_on="Sid")  # right join 
```

```python
pd.merge(df1, df2, right_index=True, left_index=True)
# inner join --> common --> right df index and left df index
```

```python
df2.index = np.arange(0, 20, 2)
```

```python
pd.merge(df1, df2, right_index=True, left_index=True)
# inner join --> common --> right df index and left df index
```

```python
pd.merge(df1, df2, right_index=True, left_index=True, how="left")
# inner join --> common --> right df index and left df index
```

```python
pd.merge(df1, df2, right_index=True, left_index=True, how="right")
# inner join --> common --> right df index and left df index
```

```python
df1.join(df2)  # left join df1 --> left, df2 --> right
# df1 and df2 --> common --> index along with the whole data of df1
```

```python
df2.join(df1)
```

```python
help(df1.join)
```

```python
df1.join(df2, on="Id", how="inner")
# df1 --> Id, df2 --> index
```
