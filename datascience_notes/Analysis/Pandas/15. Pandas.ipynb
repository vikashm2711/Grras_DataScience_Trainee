```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Datasets/LS_2.0.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
def check(x):
    v = "".join(x.split("\n")[0][3:].split(","))
    try:
        v = int(v)
    except:
        v = 0
    return v
```

```python
data['ASSETS'] = data['ASSETS'].apply(check)
```

```python
data['ASSETS']
```

```python
data[data['WINNER']==1].groupby('PARTY')['STATE'].count().nlargest(1)
```

```python
temp = data[data['WINNER']==1]
states = data[data['WINNER']==1]['STATE'].unique()
for s in states:
    v = data[data['STATE'] == s]['PARTY'].value_counts()
    print(s, "\t", v.index[0], "\t", v[0])
```
