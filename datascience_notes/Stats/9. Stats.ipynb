```python
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(data, 'o', label="MARKS")
mean = data.mean()
med = np.median(data)
plt.plot([0, 120], [mean, mean], 'r--', label="MEAN")
plt.plot([0, 120], [med, med], 'k--', label="MEDIAN")
plt.legend()
plt.show()
```

```python
cleaned_data = []
for i in data:
    if 0<=i<=100:
        cleaned_data.append(i)
cleaned_data = np.array(cleaned_data)
```

```python
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(cleaned_data, 'o', label="MARKS")
mean = cleaned_data.mean()
med = np.median(cleaned_data)
plt.plot([0, 120], [mean, mean], 'r--', label="MEAN")
plt.plot([0, 120], [med, med], 'k--', label="MEDIAN")
plt.legend()
plt.show()
```

### Whenever we don't know the range of the data then we cannot remove the values
in that case we replace the values with mean 

```python
mean = cleaned_data.mean()
replaced_data = []
for i in data:
    if 0<=i<=100:
        replaced_data.append(i)
    else:
        replaced_data.append(mean)
```

```python
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(cleaned_data, 'ko', label="MARKS")
mean = cleaned_data.mean()
plt.plot([0, 120], [mean, mean], 'r--', label="MEAN")
for i in range(len(cleaned_data)):
    plt.plot([i, i], [mean, cleaned_data[i]], '--')
plt.legend(loc="center")
plt.show()
```
