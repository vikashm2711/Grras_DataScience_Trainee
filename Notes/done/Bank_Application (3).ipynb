```python
bank = {'acc':[1001,1002,1003,1004,1005],
        'user':['Ravi','Abhi','Manmohan','Heta','Shaktimaan'],
        'pass':['python@123','kabhi@123','hello@123','hey@123','geeta@123'],
        'bal':[50.50,1000000,1204102412,4120134134,1351351334]
        }
```

```python
from getpass import getpass
print("*"*125)
print("-"*125)
print("Welcome to Python Bank".center(125))
print("-"*125)
print("*"*125)
admin = "root"
adm_pass = "redhat@123"
while True:
    print("\n\t$$--Main Menu--$$")
    print("\n\t1. Login\n\t2. Signup\n\t3. Admin Login\n\t4. Exit")
    ch = input("\nEnter your Choice --> ")
    if ch=="1":
        l_acc = int(input("\nEnter your Account Number -->"))
        if l_acc in bank['acc']:
            index = bank['acc'].index(l_acc)
            l_pass = getpass("\nEnter the Password --> ")
            if l_pass == bank['pass'][index]:
                while True:
                    print("\n\t$--Login Menu--$")
                    print("\n\t1. Credit\n\t2. Debit\n\t3. Change Password\n\t4. Check Balance\n\t5. Logout")
                    l_ch = input("\nEnter Your choice --> ")
                    if l_ch == "1":
                        c_amt = float(input("Enter the Amount You want to Credit to your Account --> "))
                        bank['bal'][index]+=c_amt
                        print(f"\nYou have Successfully Credited Rs {c_amt} to your Account\nNow the Balance in your Account is {bank['bal'][index]}")
                    elif l_ch == "2":
                        d_amt = float(input("Enter the Amount You want to Debit from your Account --> "))
                        if d_amt>bank['bal'][index]:
                            print("\nYou Have Insufficient Funds!!!")
                        else:
                            bank['bal'][index]-=d_amt
                            print(f"\nYou have Successfully Debited Rs {d_amt} from your Account\nThe remaining Balance in your Account is {bank['bal'][index]}")
                    elif l_ch == "3":
                        old_pass = getpass("\nEnter Your Current Password!!")
                        if old_pass == bank['pass'][index]:
                            n_pass = getpass("\nCreate Your New Password --> ")
                            c_pass = getpass("\nConfirm Your New Password --> ")
                            if n_pass == c_pass:
                                bank['pass'][index] = n_pass
                                print("\nYour Password has been Successfully Changed!!!")
                            else:
                                print("\nPassword Doesnot Matched!!!")
                        else:
                            print("\nIncorrect Password!!!")
                    elif l_ch=="4":
                        print(f"\nYou Account Balance is Rs {bank['bal'][index]}")
                    elif l_ch=="5":
                        c1 = input("\nDo You want to Logout? (y/n) --> ")
                        if c1=="yes" or c1 == "y":
                            print("\nYou are Successfully Logged Out!!!")
                            break
                        else:
                            continue
                    else:
                        print("\nInvalid Choice!!!")
            else:
                print("\nIncorrect Password!!!")
        else:
            print("\nInvalid Account Number!!!")
    elif ch=="2":
        print("\n$--Welcome to Signup Menu--$")
        n_user = input("\nEnter Username --> ")
        n_pass = getpass("\nCreate a Password --> ")
        n_bal = 0.0
        n_acc = bank['acc'][-1]+1
        bank['acc'].append(n_acc)
        bank['pass'].append(n_pass)
        bank['user'].append(n_user)
        bank['bal'].append(n_bal)
        print(f"\nYou have successfully SignedUp\nNow You can Login\nYour Account Number is --> {n_acc}")
    elif ch=="3":
        print("\n$--Welcome to Admin Panel--$")
        a_ = input("\nEnter Admin username --> ")
        if a_ == admin:
            a_p = getpass("\nEnter Admin Password --> ")
            if a_p == adm_pass:
                print("\nAccount Details\n")
                for var in bank:
                    print(var,"-->",bank[var])
            else:
                print("\nIncorrect Admin Password!!!")
        else:
            print("\nInvalid Admin User!!!")
    elif ch == "4":
        c = input("\nDo You want to Exit? (y/n) --> ")
        if c=="yes" or c == "y":
            print("\nThankYou For using Services!!!")
            break
        else:
            continue
    else:
        print("\nInvalid Choice!!!")
```
